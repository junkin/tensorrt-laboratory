# Mel Banks

```python
import librosa
```

```python
sample_freq = 16000
num_fft = 512
num_features = 64
mel_basis = librosa.filters.mel(sample_freq, num_fft, n_mels=num_features, fmin=0, fmax=int(sample_freq / 2))
```
